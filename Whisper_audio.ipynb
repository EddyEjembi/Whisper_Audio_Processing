```python
data_folder = '/content/drive/MyDrive/Audio_Data/121726'

file_path = os.path.join(data_folder, '121-8555.trans.txt')
with open(file_path, 'r') as f:
    lines = f.readlines()
```

```python
data = []
for line in lines:
    audio_file, transcription = line.strip().split(' ', 1)
    audio_path = os.path.join(data_folder, f'{audio_file}.flac')
    data.append((audio_path, transcription))
```

```python
from datasets import Dataset

# Convert the list to a dictionary
data_dict = {
    "audio_path": [item[0] for item in data],
    "transcription": [item[1] for item in data]
}

# Create a dataset
dataset = Dataset.from_dict(data_dict)

# Display the first few samples
print(dataset[:5])
```

```python
dataset = dataset.train_test_split(test_size=0.2)
```

```python
#Import Processor
from transformers import WhisperProcessor

processor = WhisperProcessor.from_pretrained("openai/whisper-medium", language="en", task="transcribe")
```

```python
#!pip install librosa
```

```python
import tensorflow as tf
import librosa
```

```python
from transformers import WhisperTokenizer

tokenizer = WhisperTokenizer.from_pretrained("openai/whisper-medium", language="en", task="transcribe")
```

```python
def process_data(data):
  audio_path = data['audio_path']
  waveform, sample_rate = librosa.load(audio_path, sr=16000)
  data['input'] = processor(waveform, sampling_rate=sample_rate, return_tensors="tf") .input_features
  data['target'] = tokenizer(data["transcription"]).input_ids
  #data['target'] = processor(data['transcription']).input_ids.squeeze()

  return data
```

```python
u = process_data(dataset['train'][0])
```

```python
processed_data = dataset.map(process_data, remove_columns=dataset.column_names["train"])
```
